```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you have already loaded your data into 'df'

# Calculate statistics
mean = df['Price'].mean()
median = df['Price'].median()
mode = df['Price'].mode().values[0]
std_dev = df['Price'].std()

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
print(f"Standard Deviation: {std_dev}")

# Create visualizations
plt.figure(figsize=(22, 6))
sns.countplot(x='Price', data=df, color='red')
plt.title("The Price")
plt.xlabel("Price Scores")
plt.ylabel("Count")
plt.show()

plt.figure(figsize=(10, 6))
plt.hist(df['Price'], bins=4, color='red', edgecolor='black')
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you have already loaded your data into 'df'

# Clean the 'Team' column to remove leading and trailing whitespaces
df['Team'] = df['Team'].str.strip()

# Define a custom color palette as a dictionary with team names exactly as they appear in your dataset
team_specific_palette = {
    'Royal Challengers Bangalore': 'red',
    'Gujarat Titans': 'green',
    'Delhi Capitals': 'orange',
    'Mumbai Indians': 'blue',
    'Chennai Super Kings': 'yellow',
    'Lucknow Super Giants': 'black',
    'Rajasthan Royals': 'skyblue',
    'Kolkata Knight Riders': 'purple',
    'Punjab Kings': 'brown',
    'Sunrisers Hyderabad': 'pink'
}

# Calculate money spent by each team in millions of US dollars
money_spent_by_each_team = df.groupby('Team')['Price'].sum() / 1_000_000

# Create a new DataFrame for money spent by each team
money_spent_by_each_team_df = money_spent_by_each_team.reset_index().sort_values(by='Price', ascending=True)

# Create a violinplot
ax = sns.violinplot(x='Price', y='Team', data=df, palette=team_specific_palette, inner="quartile")
plt.title('Distribution of Money spent by each team (US$)', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel('Money spent (Million US$)', fontsize=14)
plt.ylabel('Team', fontsize=14)
plt.show()
```

```python
fig = plt.figure(figsize=(20, 6))
ax = sns.boxplot(x='Nationality', y='Price', data=df)
plt.title('Price Distribution of Players from different Nationality', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.ylabel('Price (US$)', fontsize=14)
plt.xlabel('National Side', fontsize=14)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
mean_prices = df.groupby('Nationality')['Price'].mean().reset_index()
# Create a line plot
plt.figure(figsize=(12, 6))
sns.lineplot(x='Nationality', y='Price', data=mean_prices,marker='o', color='b')
plt.title('Mean Price (US$) of Players from Different Nationality', fontsize=16)
plt.xticks(rotation=45)
plt.xlabel('National Side', fontsize=14)
plt.ylabel('Mean Price (US$)', fontsize=14)
plt.grid(True)
plt.show()
```

```python
fig = plt.figure(figsize=(15, 6))
df['Price Label'] =df['Price'].apply(lambda x: f'${x}')
top_5_players_by_price =df.sort_values(by='Price',ascending=False).head(5)
ax = sns.barplot(y='Name', x='Price',data=top_5_players_by_price.sort_values(by='Price',ascending=True))
for container in ax.containers:ax.bar_label(container, fmt='%d', label_type='edge',fontsize=12, color='black', padding=3, weight='bold',labels=top_5_players_by_price.sort_values(by='Price',ascending=True)
['Price Label'])
plt.title('Top 5 Players based on Price (US$)', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.ylabel('')
plt.xlabel('Money Spent (in US$)', fontsize=14)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Assuming you have already loaded your data into 'df'
# Calculate the count of players for each team
team_counts = df['Team'].value_counts()
# Extract the team names and counts
teams = team_counts.index
player_counts = team_counts.values
# Create a pie chart to visualize the distribution of players among teams
plt.figure(figsize=(8, 8))
plt.pie(player_counts, labels=teams, autopct='%1.1f%%',
startangle=140)
plt.title('Distribution of Players Among Teams',
fontsize=16)
plt.show()
```
